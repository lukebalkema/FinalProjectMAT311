```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set aesthetic parameters for seaborn
sns.set(style="whitegrid")

# Define the column names as per the dataset description
column_names = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num',
    'marital-status', 'occupation', 'relationship', 'race', 'sex',
    'capital-gain', 'capital-loss', 'hours-per-week', 'native-country',
    'income'
]

# Read the combined dataset
df_total = pd.read_csv('adult.totalData.csv', header=None, names=column_names, skipinitialspace=True)

# Display the first few rows to verify
print("First 5 rows of the dataset:")
print(df_total.head())

# Check if the first row contains header names as data
if df_total.iloc[0]['age'] == 'age':
    print("\nHeader row detected as data. Removing the first row.")
    df_total = df_total.drop(0).reset_index(drop=True)
    print("First 5 rows after removing header row as data:")
    print(df_total.head())
else:
    print("\nNo header row detected in the data.")

# Replace '?' with NaN
df_total.replace('?', np.nan, inplace=True)

# Check the number of missing values per column
print("\nMissing values per column before imputation:")
print(df_total.isnull().sum())

# List of categorical columns with missing values
categorical_columns = ['workclass', 'occupation', 'native-country']

# Fill missing values with the mode of each column
for col in categorical_columns:
    mode = df_total[col].mode()[0]
    df_total[col].fillna(mode, inplace=True)

# Verify that there are no missing values left
print("\nMissing values after imputation:")
print(df_total.isnull().sum())

# Convert numerical columns to appropriate data types (if not already)
numerical_columns = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
for col in numerical_columns:
    df_total[col] = pd.to_numeric(df_total[col], errors='coerce')

# Verify data types
print("\nData types of numerical columns:")
print(df_total[numerical_columns].dtypes)

# Check for any remaining non-numeric values in numerical columns
for col in numerical_columns:
    non_numeric = df_total[col].isnull().sum()
    if non_numeric > 0:
        print(f"\nNon-numeric entries found in {col}: {non_numeric}")
    else:
        print(f"\nAll entries in {col} are numeric.")

# -----------------------------
# Exploratory Data Analysis (EDA)
# -----------------------------

# a. Distribution of the Target Variable ('income')
plt.figure(figsize=(6,4))
sns.countplot(x='income', data=df_total, palette='Set2')
plt.title('Distribution of Income')
plt.xlabel('Income')
plt.ylabel('Count')
plt.xticks([0,1], ['<=50K', '>50K'])
plt.show()

# b. Numerical Features Analysis

# i. Histograms
df_total[numerical_columns].hist(bins=30, figsize=(15,10), layout=(3,2))
plt.suptitle('Histograms of Numerical Features')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to accommodate suptitle
plt.show()

# ii. Boxplots
plt.figure(figsize=(15,10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 2, i)
    sns.boxplot(y=col, data=df_total, palette='Set3')
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()

# iii. Correlation Heatmap
corr_matrix = df_total[numerical_columns].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()

# c. Categorical Features Analysis

# i. Count Plots for Categorical Features
plt.figure(figsize=(20, 25))
for i, col in enumerate(categorical_columns, 1):
    plt.subplot(4, 2, i)
    sns.countplot(y=col, data=df_total, palette='Set2', order=df_total[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
plt.tight_layout()
plt.show()

# ii. Bar Plots of Categorical Features vs. 'income'
plt.figure(figsize=(20, 25))
for i, col in enumerate(categorical_columns, 1):
    plt.subplot(4, 2, i)
    sns.countplot(y=col, hue='income', data=df_total, palette='Set2')
    plt.title(f'{col} vs. Income')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.legend(title='Income', labels=['<=50K', '>50K'])
plt.tight_layout()
plt.show()

# iii. Boxplots of Numerical Features vs. 'income'
plt.figure(figsize=(15,10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 2, i)
    sns.boxplot(x='income', y=col, data=df_total, palette='Set3')
    plt.title(f'{col} vs. Income')
    plt.xlabel('Income')
    plt.ylabel(col)
plt.tight_layout()
plt.show()

# d. Pivot Tables

# Example: Average age by income
age_income = df_total.pivot_table(index='income', values='age', aggfunc='mean')
print("\nAverage Age by Income:")
print(age_income)

# Example: Average hours-per-week by income
hours_income = df_total.pivot_table(index='income', values='hours-per-week', aggfunc='mean')
print("\nAverage Hours per Week by Income:")
print(hours_income)

# Example: Count of each workclass by income
workclass_income = pd.crosstab(df_total['workclass'], df_total['income'])
print("\nWorkclass Counts by Income:")
print(workclass_income)

# -----------------------------
# Visualization Enhancements
# -----------------------------

# i. Heatmap of Education vs. Income
education_income = pd.crosstab(df_total['education'], df_total['income'])
plt.figure(figsize=(12,8))
sns.heatmap(education_income, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap of Education vs. Income')
plt.xlabel('Income')
plt.ylabel('Education')
plt.xticks([0,1], ['<=50K', '>50K'])
plt.show()

# ii. Pair Plot for Numerical Features Colored by Income
# Sample a subset for pairplot to reduce computational load
sample_df = df_total.sample(n=1000, random_state=123)

sns.pairplot(sample_df, vars=['age', 'education-num', 'hours-per-week', 'capital-gain', 'capital-loss'], hue='income', palette='Set2', diag_kind='kde')
plt.suptitle('Pairplot of Numerical Features Colored by Income', y=1.02)
plt.show()

# iii. Feature Importance (Using Decision Tree if applicable)
# Assuming you have trained a Decision Tree classifier named 'best_model' with max_depth=10
# Replace 'best_model' with your actual model variable if different

# Uncomment the following section if 'best_model' exists

# try:
#     feature_importances = pd.Series(best_model.feature_importances_, index=X_train.columns)
#     top_features = feature_importances.sort_values(ascending=False).head(20)
    
#     # Plot feature importances
#     plt.figure(figsize=(10,8))
#     sns.barplot(x=top_features.values, y=top_features.index, palette='viridis')
#     plt.title('Top 20 Feature Importances from Decision Tree')
#     plt.xlabel('Importance Score')
#     plt.ylabel('Features')
#     plt.show()
# except NameError:
#     print("\nError: 'best_model' is not defined. Please ensure you've trained the Decision Tree classifier and named it 'best_model'.")

# -----------------------------
# Summary Tables
# -----------------------------

# Summary statistics for numerical features
print("\nSummary Statistics for Numerical Features:")
print(df_total[numerical_columns].describe())

# Frequency counts for categorical features
for col in categorical_columns:
    print(f"\nFrequency Counts for {col}:")
    print(df_total[col].value_counts())

# -----------------------------
# Additional Insights
# -----------------------------

# i. Income Distribution Across Different Countries (Top 10 Countries by Count)
top_countries = df_total['native-country'].value_counts().nlargest(10).index
country_income = df_total[df_total['native-country'].isin(top_countries)]

plt.figure(figsize=(12,8))
sns.countplot(y='native-country', hue='income', data=country_income, palette='Set2')
plt.title('Income Distribution Across Top 10 Countries')
plt.xlabel('Count')
plt.ylabel('Native Country')
plt.legend(title='Income', labels=['<=50K', '>50K'])
plt.show()

# ii. Education Level vs. Income
plt.figure(figsize=(12,8))
sns.countplot(y='education', hue='income', data=df_total, palette='Set2', order=df_total['education'].value_counts().index)
plt.title('Income Distribution Across Education Levels')
plt.xlabel('Count')
plt.ylabel('Education')
plt.legend(title='Income', labels=['<=50K', '>50K'])
plt.show()
```
